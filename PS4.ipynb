```python
import pandas as pd
import numpy as np

%matplotlib inline

HEALTH_CONVERSION_RULE = {"Poor" : 0, "Fair": 1, "Good" : 2, "Very good": 3, "Excellent" : 4}

RACE_CONVERSION_RULE = {"White": 0, "Black": 1}
```

```python
df = pd.read_stata("longdata.dta")
df.columns
```

```python
df["quantized_health"] = df.apply(lambda x: HEALTH_CONVERSION_RULE[x.health] if x.health == x.health else x.health, axis = 1)

df.replace(to_replace = "Other", value = np.nan, inplace = True)
df["quantized_race"] = df.apply(lambda x: RACE_CONVERSION_RULE[x.race] if x.race == x.race else x.race, axis = 1)


df.dropna(axis = 0, subset = ["quantized_health", "quantized_race", "educ"], inplace=True)
df = df[['famid', 'relation', 'educ', 'age', 'quantized_race', 'quantized_health']]
```

## The answer to question #2

```python
parents = df[df.relation != "Child"].groupby(["famid"]).mean()

children = df[df.relation == "Child"].groupby(["famid"]).mean()

parents_h = parents[['quantized_health']]
parents_h.columns = ['parent_health']

regression_ready = children.join(parents_h, how = "inner")
regression_ready['const'] = np.ones(len(regression_ready))
regression_ready
```

```python
from sklearn import linear_model
```

```python
reg_a = linear_model.LinearRegression()
reg_a.fit(regression_ready[["parent_health", "const"]], regression_ready["quantized_health"])

reg_b = linear_model.LinearRegression()
reg_b.fit(regression_ready[["parent_health", "const", "age"]], regression_ready["quantized_health"])

reg_c = linear_model.LinearRegression()
reg_c.fit(regression_ready[["parent_health", "const", "age", "educ"]], regression_ready["quantized_health"])

reg_d = linear_model.LinearRegression()
reg_d.fit(regression_ready[["parent_health", "const", "age", "quantized_race"]], regression_ready["quantized_health"])
```

```python
print ("The answers:")
print ("1: " + str(len(regression_ready)))
print ("2: See above")
print ("3a: " + str(reg_a.coef_[0]))
print ("3b: " + str(reg_b.coef_[0]))
print ("3c: " + str(reg_c.coef_[0]))
print ("3d: " + str(reg_d.coef_[0]))
```
